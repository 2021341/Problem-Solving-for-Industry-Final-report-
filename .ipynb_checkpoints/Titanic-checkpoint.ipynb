We drop some of the columns as they don't seem to have relevance in the survival or not of the passanger (Name, Ticket Number, Passenger Flare)  or they are missing too much data like 'Cabin'.

```python
titanic_data1=titanic_data[['Passenger Class','Gender','Age','No of Siblings','No of Parents' ,'Passenger Fare',
                          'Port of Embarkation','Survived']]
```

```python
titanic_data1.head()
```

We need to convert features into numeric ones to work with KNN algorithm.

We describe Gender, Passenger Class and Survived columns with numerical values. We replace the strings and then make the column numeric.

```python
titanic_data1=titanic_data1.replace(to_replace="Female", value=1)
titanic_data1=titanic_data1.replace(to_replace="Yes", value=1)
titanic_data1=titanic_data1.replace(to_replace="Male", value=0)
titanic_data1=titanic_data1.replace(to_replace="No", value=0)
titanic_data1=titanic_data1.replace(to_replace="First", value=1)
titanic_data1=titanic_data1.replace(to_replace="Second", value=2)
titanic_data1=titanic_data1.replace(to_replace="Third", value=3)
titanic_data1["Gender"]=titanic_data1["Gender"].apply(pd.to_numeric)
titanic_data1["Survived"]=titanic_data1["Survived"].apply(pd.to_numeric)
titanic_data1["Passenger Class"]=titanic_data1["Passenger Class"].apply(pd.to_numeric)
```

```python
titanic_data1.head()
```

```python
titanic_data1.info()
```

```python
titanic_data1.to_csv("Titanic_Mod.csv")
```

```python
data_port=pd.read_csv("Titanic_Mod.csv",usecols=['Port of Embarkation'])
```

```python
titanic_num=pd.concat([titanic_data1[['Passenger Class','Gender','Age','No of Siblings','No of Parents',
                   'Passenger Fare',]],pd.get_dummies(data_port), titanic_data1[['Survived']]], axis=1) 
```

```python
titanic_num.head()
```

```python
titanic_num.iloc[[168,284],[6,7,8]]
```

```python
titanic_num.info()
```

Now we have our values in numerical format to apply KNN. Our y will be the "Survived" column and X will be our dataset after dropping the "Survived" column

```python
y=titanic_num[['Survived']]
```

```python
X=titanic_num[['Passenger Class','Gender',
             'Port of Embarkation_Cherbourg','Port of Embarkation_Queenstown','Port of Embarkation_Southampton']]
```

```python
y.dtypes
```

```python
X.dtypes
```

```python
X.head()
```

```python
X.to_csv("X.csv")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
error_rate = []
# Will take some time
for i in range(1,50):
 knn = KNeighborsClassifier(n_neighbors=i, metric='euclidean')
 knn.fit(X_train,y_train)
 pred_i = knn.predict(X_test)
 error_rate.append(np.mean(pred_i != [y_test]))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50),error_rate,color="blue", linestyle="dashed", marker="o",
 markerfacecolor="red", markersize=10)
plt.title("Error Rate vs. K Value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```

```python
knn = KNeighborsClassifier(n_neighbors=2, metric='euclidean')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
```

(191+66)/(191+66+4+67)=0.7835 So our Model is 78.4% accurate

# Given our confustion matrix, our model has an accuracy of 78.4%
