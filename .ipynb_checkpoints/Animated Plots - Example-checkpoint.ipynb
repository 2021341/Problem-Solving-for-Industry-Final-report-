```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
df = pd.read_csv("share-of-individuals-using-the-internet.csv")
```

```python
df.head()
```

```python
br = df[df["Country"] == "Brazil"]
```

```python
br.reset_index(inplace=True)
```

```python
br.shape
```

```python
ire = df[ df["Country"] == "Ireland"]
```

```python
ire.reset_index(inplace=True)
```

```python
ire.shape
```

```python
ire["Year"].min()
```

```python
%matplotlib notebook
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
fig, ax = plt.subplots()
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1, 1)
line, = ax.plot([], [])
def update(frame):
    x = np.linspace(0, 2*np.pi, 128)
    y = np.sin(x + frame/10.0)
    line.set_data((x, y))
    return line,
ani = FuncAnimation(fig, update, frames=range(100), blit=True)
plt.show()
```

```python
%matplotlib
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)

ax.set_xlim(ire["Year"].min(), ire["Year"].max())
ax.set_ylim(0, 100)

line1 = plt.plot(ire["Individuals using the Internet (% of population)"][0], color="green")[0]
line2 = plt.plot(br["Individuals using the Internet (% of population)"][0], color="blue")[0]

point1 = plt.plot(ire["Individuals using the Internet (% of population)"][0], color="green", marker="o")[0]
point2 = plt.plot(br["Individuals using the Internet (% of population)"][0], color="blue", marker="o")[0]
def update(frame):
    line1.set(xdata=ire["Year"][:frame+1].to_numpy(),ydata=ire["Individuals using the Internet (% of population)"][:frame+1].to_numpy())
    line2.set(xdata=ire["Year"][:frame+1].to_numpy(),ydata=br["Individuals using the Internet (% of population)"][:frame+1].to_numpy())
    
    point1.set(xdata=ire["Year"][frame],ydata=ire["Individuals using the Internet (% of population)"][frame])
    point2.set(xdata=ire["Year"][frame],ydata=br["Individuals using the Internet (% of population)"][frame])
    return line1, line2

ax.legend(["Ireland", "Brazil"])

ani = animation.FuncAnimation(fig=fig, func=update, frames=br.shape[0], interval=500)
plt.show()
```
