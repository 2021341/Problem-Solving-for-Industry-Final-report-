## Dashboard With GUI - TKinter

```python
import tkinter as tk
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

```python
root = tk.Tk()

# Plotting df1 as a bar chart
fig1 = plt.Figure(figsize=(6, 5), dpi=100)
ax1 = fig1.add_subplot(111)

bar1 = FigureCanvasTkAgg(fig1, root)
bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)

df1.plot(kind="bar", legend=True, ax=ax1)
ax1.set_title("Country vs GDP Per Capita")

# Plotting df2 as a line chart/time series plot
fig2 = plt.Figure(figsize=(6, 5), dpi=100)
ax2 = fig2.add_subplot(111)

line2 = FigureCanvasTkAgg(fig2, root)
line2.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)

df2.plot(kind="line", legend=True, ax=ax2, color="r", marker="o")
ax2.set_title("Unemployment Rate from 1920 to 2010")

# Plotting df3 as a scatter plot
# df3.plot(kind="scatter")
# ax3.scatter(x=col1, y=col2)

fig3 = plt.Figure(figsize=(6, 5), dpi=100)
ax3 = fig3.add_subplot(111)

scatter3 = FigureCanvasTkAgg(fig3, root)
scatter3.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)

df3.plot(kind="scatter", x="Interest_rate", y="Stock_index_price", legend=True, ax=ax3, color="g")
ax3.set_title("Stock Index Price vs Interest Rate")

root.mainloop()
```

## Adding Interactivity

```python
from tkinter import ttk
```

```python
root = tk.Tk()

canvas = tk.Canvas(root, width=800, height=300)

label = tk.Label(root, text="GUI")
label.config(font=("Ariel", 20))
canvas.create_window(400, 50, window=label)

# Dropdown menu
combobox = ttk.Combobox(root)
combobox["values"] = ("df1", "df2", "df3")
canvas.create_window(400, 220, window=combobox)

combobox.grid(row=0, column=1)
combobox.current(1)

def createCharts():
    clearCharts()
    global chart
    grid_position = [4, 3]

    # Get the data that the user selected
    df = combobox.get()

    if df == "df1":
        # Plotting df1 as a bar chart
        fig1 = plt.Figure(figsize=(6, 5), dpi=100)
        ax1 = fig1.add_subplot(111)
        
        chart = FigureCanvasTkAgg(fig1, root)
        chart.get_tk_widget().grid(row=grid_position[0], column=grid_position[1])
        
        df1.plot(kind="bar", legend=True, ax=ax1)
        ax1.set_title("Country vs GDP Per Capita")
    elif df == "df2":
        # Plotting df2 as a line chart/time series plot
        fig2 = plt.Figure(figsize=(6, 5), dpi=100)
        ax2 = fig2.add_subplot(111)
        
        chart = FigureCanvasTkAgg(fig2, root)
        chart.get_tk_widget().grid(row=grid_position[0], column=grid_position[1])
        
        df2.plot(kind="line", legend=True, ax=ax2, color="r", marker="o")
        ax2.set_title("Unemployment Rate from 1920 to 2010")
    else:
        # Plotting df3 as a scatter plot
        fig3 = plt.Figure(figsize=(6, 5), dpi=100)
        ax3 = fig3.add_subplot(111)
        
        chart = FigureCanvasTkAgg(fig3, root)
        chart.get_tk_widget().grid(row=grid_position[0], column=grid_position[1])
        
        df3.plot(kind="scatter", x="Interest_rate", y="Stock_index_price", legend=True, ax=ax3, color="g")
        ax3.set_title("Stock Index Price vs Interest Rate")

def clearCharts():
    try: # Get rid of old chart, if exists
        chart.get_tk_widget().grid_forget()
    except Exception as e: # Do nothing if no chart exists
        print(e)
        pass

# Buttons
createChartsButton = tk.Button(
    root,
    text="Create Chart",
    command=createCharts,
    bg="palegreen", # button background colour
    font=("Poppins", 11, "bold")
).grid(row=1, column=0)
canvas.create_window(400, 100, window=createChartsButton)

clearChartsButton = tk.Button(
    root,
    text="Clear Chart",
    command=clearCharts,
    bg="yellow", # button background colour
    font=("Poppins", 11, "bold")
).grid(row=2, column=0)
canvas.create_window(400, 200, window=clearChartsButton)

exitButton = tk.Button(
    root,
    text="Exit",
    command=root.destroy,
    bg="red", # button background colour
    font=("Poppins", 25, "bold")
).grid(row=3, column=0)
canvas.create_window(600, 200, window=exitButton)

root.mainloop()
```
