```python
table = { 
    'Hello' : ['Jane', 'Paul'],
    'Jane' : [],
    'Paul' : ['Simon'],
    'Simon' : ['Ann', 'Oliver'],
    'Ann' : ['Tito'],
    'Oliver' : [],
    'Tito' : []    
}
```

```python
visited = set()
```

```python
def dfs(visited, table, node):
    if node not in visited:
        print (node)
        visited.add(node)
        for neighbour in table[node]:
            dfs(visited, table, neighbour)
```

```python
dfs(visited, table, 'Hello')
```
