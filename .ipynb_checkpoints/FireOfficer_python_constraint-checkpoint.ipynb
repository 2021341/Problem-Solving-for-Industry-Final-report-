```python
import constraint
problem = constraint.Problem()

fire_officers=["Phylis", "Ann", "Henry", "Eva", "Bill", "Mark", "Bob"]

problem.addVariables(fire_officers,range(1,5))

def phylis_1(phylis):
    return phylis == 1

problem.addConstraint(phylis_1, ["Phylis"])

def phylis_eva(phylis, eva):
    return phylis != eva

problem.addConstraint(phylis_eva, ["Phylis", "Eva"])

def mark_bob(mark, bob):
    return mark == bob

problem.addConstraint(mark_bob, ["Mark", "Bob"])

def henry_eva(henry, eva):
    return henry == eva

problem.addConstraint(henry_eva, ["Henry", "Eva"])

def eva_mark_bob_bill(eva, mark, bob, bill):
    return eva != mark and eva != bob and eva != bill

problem.addConstraint(eva_mark_bob_bill, ["Eva", "Mark", "Bob", "Bill"])

def ann_not_adjacent(ann, phylis, eva):
    return (
        ann != phylis and ann != eva and
        abs(ann - phylis) != 1 and abs(ann - eva) != 1
    )

problem.addConstraint(ann_not_adjacent, ["Ann", "Phylis", "Eva"])



solutions = problem.getSolutions()
print(solutions)
print(len(solutions))

```
