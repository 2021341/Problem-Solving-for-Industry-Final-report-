## KNN

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set()
```

```python
data_df = pd.read_csv("breast_cancer.csv")
```

```python
data_df.head()
```

```python
data_df.info()
```

```python
data_df.describe()
```

```python
sns.scatterplot(
    x='area_mean',
    y='compactness_mean',
    data=data_df
)
```

```python
X=data_df.drop(columns=["id","Unnamed: 32","diagnosis"])
```

```python
X.head()
```

```python
y = data_df["diagnosis"]
y = pd.get_dummies(y, drop_first=True)
```

```python
y.tail()
```

```python
y = y.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
knn = KNeighborsClassifier(n_neighbors=7, metric='euclidean')
knn.fit(X_train, y_train.ravel())
```

```python
y_pred = knn.predict(X_test)
```

```python
plt.scatter(
    X_test['area_mean'],
    X_test['compactness_mean'],
    c=y_pred,
    cmap='coolwarm',
    alpha=0.7
)
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

### Finding the Best Value for "K"
Lets create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.

```python
error_rate = []
# Will take some time
for i in range(1,40):
 
 knn = KNeighborsClassifier(n_neighbors=i, metric='euclidean')
 knn.fit(X_train,y_train.ravel())
 pred_i = knn.predict(X_test)
 error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color="blue", linestyle="dashed", marker="o",
 markerfacecolor="red", markersize=10)
plt.title("Error Rate vs. K Value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```

The optimal value of K is 2. The graph of the error rate increases after 2.
