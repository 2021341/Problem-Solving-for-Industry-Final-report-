```python
from joblib import Parallel, delayed 
import joblib 

import warnings
warnings.filterwarnings('ignore')
```

### Load the model from the file 

```python
knn_from_joblib = joblib.load('TitanicKnn.pkl') 
```

I now need some data to make a prediction with. This meens I need a value for Passenger Class (Integer between 1-3),Gender(Integer between 0-1),Port of Embarkation_Cherbourg(Boolean),Port of Embarkation_Queenstown(Boolean),Port of Embarkation_Southampton(Boolean).

So a first class female passenger who embarked from Southampton would be represented as 1,1,False,False,True (Remember that SKLearn KNN need a 2D array for input so we will express it as a 2d array: [[1,1,False,False,True]] )

```python
new_data=[[1,1,False,False,True]]
```

We then just pass the data into the Trained model to make a prediction. If we get a 1 the model has predicted that the passenger will survive, if we get a 0 then the prediction is that the passenger will not survive 

```python
knn_from_joblib.predict(new_data)
```

We got a 1, so the model has predicted that the passenger will survive! (Remember however that the model is only 78.4% accurate)
