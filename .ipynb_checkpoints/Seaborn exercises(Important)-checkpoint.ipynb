```python
# 1. Import necessary modules
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt  # Adding matplotlib to display the plot

# 2. Load the diamonds dataset from Seaborn
diamonds_df = sns.load_dataset('diamonds')

# 3. Plot a histogram using Seaborn's histplot function with carat on the x-axis
sns.histplot(data=diamonds_df, x='carat', bins=30, kde=True)  # kde=True adds a density curve
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.title('Distribution of Diamond Carat Weights')
plt.show()
```
