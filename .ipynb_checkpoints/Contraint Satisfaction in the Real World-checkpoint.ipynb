## Contraint Satisfaction in the Real World

Google Optimization Tools (a.k.a., OR-Tools) is an open-source software suite for solving combinatorial optimization problems. Our problem will be modeled using this framework in Python.

```python
#pip install ortools
```

```python
import numpy as np
from random import randint
from math import ceil
import matplotlib.pyplot as plt # Data visualization
plt.rcParams["figure.figsize"] = (15,10)
from ortools.sat.python import cp_model # CP-SAT solver
```

#### Parameters

For now, let’s simplify the problem to 4 parameters (1):
 1. Location of infected people
 2. Severity of infected people
 3. Location of hospitals
 4. Number of beds in each hospital
 
Let’s define those parameters 


Number of hospitals

```python
n_hospitals = 3
```

Number of infected people

```python
n_patients = 200
```

Number of beds in every hospital

```python
n_beds_in_hospitals = [30,50,20]
```

Location of infected people -- random integer tuple (x,y)

```python
patients_loc = [(randint(0, 100), randint(0, 100)) for _ in range(n_patients)]
```

Location of hospitals -- random integer tuple (x,y)

```python
hospitals_loc = [(randint(0, 100), randint(0, 100)) for _ in range(n_hospitals)]  
```

Illness severity -- 1 = mild -> 5 = severe

```python
patients_severity = [randint(1, 5) for _ in range(n_patients)]
```

### Visualization

```python
shade_severity = [int(sev*255/5) for sev in patients_severity]
color_severity = ["#%02x0000" % (sev) for sev in shade_severity]
size_severity = [k**2.5 for k in patients_severity]
plt.scatter(*zip(*patients_loc), s=size_severity, c=color_severity, label="Patients")
plt.scatter(*zip(*hospitals_loc), s=500, c="g", marker="P", label="Hospitals")
plt.legend()
plt.axis('off')
plt.show()       
```

### Variables

A constraint satisfaction problem consists of a set of variables that must be assigned values in such a way that a set of constraints is satisfied.

 1. Let I be the set of hospitals
 2. Let Jᵢ be the set of beds in hospital i
 3. Let K be the set of patients.

Let define as our indexed family of variables :


If in the hospital i, the bed j is taken by the person k then xᵢⱼₖ = 1. In order to associate each bed of an hospital to an ill person, the goal is to find a set of variables that satisfies all of our constraints.

```python
model = cp_model.CpModel()
x = {}
for i in range(n_hospitals):
  for j in range(n_beds_in_hospitals[i]):
    for k in range(n_patients):
      x[(i,j,k)] = model.NewBoolVar("x(%d,%d,%d)" % (i,j,k))
```

### Hard constraints

Hard constraints define a goal for our model. They are essential, if they are not resolved then the problem can’t be tackled:

 1. There must be at most a single person in every bed,
 2. There must be at most a single bed assigned to every person.

Let’s focus on the first hard constraint. For each bed j in every hospital i:

 1. Either there is a unique patient k,
 2. Either the bed is empty.


Each bed must host at most one person

```python
for i in range(n_hospitals):
  for j in range(n_beds_in_hospitals[i]):
    model.Add(sum(x[(i,j,k)] for k in range(n_patients)) <= 1)
```

Next, the second hard constraint: for every patient k:
 1. Either he is in a unique bed j in a unique hospital i,
 2. OR he is at home.


Each person must be placed in at most one bed

```python
for k in range(n_patients):
  inner_sum = []
  for i in range(n_hospitals):
    inner_sum.append(sum(x[(i,j,k)] for j in range(n_beds_in_hospitals[i]))) 
  model.Add(sum(inner_sum) <= 1)
```

### Soft constraints
Next, there are soft constraints. Those are highly desired: our solution must try to satisfy them as much as possible, yet they are not essential to find a solution:
 1. Every sick person should be placed into a bed,
 2. Every person should be handled by the nearest hospital,
 3. Sick persons in severe condition should be handled first when there are not enough beds.


Integer distance function

```python
idist = lambda xy1, xy2: int(((xy1[0]-xy2[0])**2 + (xy1[1]-xy2[1])**2)**0.5)# Gain factors (1/penalty factors)
gain_max_patients = 140
gain_severity = int(140/5)
gain_distance = -1# Maximization objective
soft_csts = []
for i in range(n_hospitals):
  for j in range(n_beds_in_hospitals[i]):
    for k in range(n_patients):
      factor = \
        gain_max_patients \
        + gain_distance * idist(hospitals_loc[i], patients_loc[k]) \
        + gain_severity * patients_severity[k]
      soft_csts.append(factor * x[(i,j,k)])
model.Maximize(sum(soft_csts))
```

### Solver
Now we can launch the solver. It will try to find the optimal solution within a specified time limit. If it can’t manage to find the optimal solution, it will return the closest sub-optimal solution.


```python
solver = cp_model.CpSolver()
solver.parameters.max_time_in_seconds = 60.0
status = solver.Solve(model)
```

### Results

```python
plot_lines = [[] for _ in range(n_hospitals)]
for i in range(n_hospitals):
  print("Hospital ", i)
  for j in range(n_beds_in_hospitals[i]):
    for k in range(n_patients):
      if solver.Value(x[(i,j,k)]) == 1:
        print("\tPatient ", k, "in bed ", j)
        line_abs = [ hospitals_loc[i][0], patients_loc[k][0] ]
        line_ord = [ hospitals_loc[i][1], patients_loc[k][1] ]
        plot_lines[i].append([line_abs, line_ord])
print()
print("Inference time: %f s" % solver.WallTime())
```

```python
from itertools import cycle

shade_severity = [int(sev*255/5) for sev in patients_severity]
color_severity = ["#%02x0000" % (sev) for sev in shade_severity]
size_severity = [k**2.5 for k in patients_severity]
plt.scatter(*zip(*patients_loc), s=size_severity, c=color_severity, label="Patients", zorder=2)
plt.scatter(*zip(*hospitals_loc), s=500, c="g", marker="P", label="Hospitals", zorder=3)
colors = cycle('bgrcmk')
for i in range(n_hospitals):
  c = next(colors)
  for (x_,y_) in plot_lines[i]:
    plt.plot(x_,y_,c=c,linewidth=0.2, zorder=1)
plt.legend()
plt.axis('off')
plt.show()       
```
